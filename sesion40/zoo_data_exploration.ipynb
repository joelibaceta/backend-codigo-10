```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
zoo_df = pd.read_csv('zoo.csv')
```

```python
class_df = pd.read_csv('class.csv')
```

```python
class_df.head()
```

```python
animal_df = zoo_df.merge(class_df, how="left", left_on="class_type", right_on="Class_Number")
```

```python
animal_df.head()
```

```python
animals_filtered_df = animal_df.drop(['class_type', 'Animal_Names', 'Number_Of_Animal_Species_In_Class'], axis=1)
```

```python
animals_filtered_df.head()
```

```python
animals_filtered_df.info()
```

```python
animals_filtered_df.describe()
```

```python
plt.rcParams['figure.figsize'] = (7,7)
```

```python
sns.countplot(x=animals_filtered_df['Class_Type'])
```

```python
plt.rcParams['figure.figsize'] = (16,10)
```

```python
df_temp = animals_filtered_df.drop(['Class_Number'], axis=1)
```

```python
df_temp = df_temp.groupby(by='animal_name').mean()
```

```python
sns.heatmap(df_temp)
```
